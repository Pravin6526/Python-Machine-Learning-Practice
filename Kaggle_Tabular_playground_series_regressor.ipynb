```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
train.shape
```

```python
train.info()
```

```python
test = pd.read_csv("test.csv")
test
```

```python
submission =pd.read_csv("sample_submission.csv")
submission
```

```python
#Define target and drop from train
target = train[['target_carbon_monoxide', 'target_benzene', 'target_nitrogen_oxides']]
train.drop(['target_carbon_monoxide', 'target_benzene', 'target_nitrogen_oxides'], axis = 1,inplace = True)
train
```

```python
#Combine train and test
combi = train.append(test)
combi
```

```python
#Drop date_time
combi.drop(['date_time'], axis = 1, inplace = True)
combi
```

```python
#Check for null values
combi.isnull().sum()
```

```python
#Normalise combi
combi = (combi.max() - combi) / (combi.max() - combi.min())
combi
```

```python
#Split X for training and validationg
y = target
X = combi[: len(train)]
X_test = combi[len(train) :]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.10, random_state=1)
X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape
```

```python
#Define model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
r2_score = reg.score(X_train,y_train)
print("Accuracy :- ",r2_score*100,'%')
```

```python
#Accuracy of validation sets
reg.fit(X_val,y_val)
r_score = reg.score(X_val,y_val)
print("Accuracy :- ",r_score*100,'%')
```

```python
pred1=reg.predict(X_val)
pred1
```

```python
pred=reg.predict(X_test)
pred
```

```python
submission.target_carbon_monoxide = pred[:, 0]
submission.target_benzene = pred[:, 1]
submission.target_nitrogen_oxides = pred[:, 2]
submission.to_csv('submission1.csv', index=False)
submission = pd.read_csv("submission1.csv")
submission
```

#random forest Regressor

```python
#Split X for training and validationg
y = target
X = combi[ : len(train)]
X_test = combi[len(train) :]
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=42)
X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(X_train,y_train);
Acc=rf.score(X_train,y_train)
print("Accuracy :-",Acc*100,"%")
```

```python
#prediction on validation set
prediction = rf.predict(X_val)
# Calculate the absolute errors
errors = abs(prediction - y_val)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
prediction
```

```python
# Use the forest's predict method on the test data
pred1 = rf.predict(X_test)
pred1
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / y_val)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
#Prepare submission
submission.target_carbon_monoxide = pred1[:, 0]
submission.target_benzene = pred1[:, 1]
submission.target_nitrogen_oxides = pred1[:, 2]
submission.to_csv('submission.csv', index=False)
submission = pd.read_csv("submission.csv")
submission
```
