```python
def stastical_data(path,col_name):
    df=pd.read_csv(path)
    df.head()
    print(col_name)
    df.columns=[col_name]
    df.head()
    df = pd.DataFrame(df[col_name].str.split('\t',2).tolist())
    df.dropna(inplace=True)
    df[0]=pd.to_datetime(df[0])
    df.index=df[0]
    df.index=df.index.date
    df.drop(columns=[0],inplace=True)
    df.columns=[col_name]
    df.isnull().sum()
    df=df.astype({col_name:float,})
    df.info()
    final_df=df.groupby(df.index).agg({col_name:['min','max','sum','mean','median','std','var','mad','prod','count']})
    final_df
    header = np.array(['Min','Max','Sum','Mean','Median', 'Standard deviation', 'Variance', 'Mean absolute deviation','Product','Count'])
    final_df.columns=header
    print("Pre-Processing is done")
    fig=go.Figure()
    fig.add_trace(go.Scatter(x=final_df.index,y=final_df['Max'],mode='lines',name=col_name+' (Max)'))
    fig.add_trace(go.Scatter(x=final_df.index,y=final_df['Min'],mode='lines',name=col_name+' (Min)'))
    fig.add_trace(go.Scatter(x=final_df.index,y=final_df['Sum'],mode='lines',name=col_name+' (Sum)'))
    fig.add_trace(go.Scatter(x=final_df.index,y=final_df['Mean'],mode='lines',name=col_name+' (Mean)'))
    fig.add_trace(go.Scatter(x=final_df.index,y=final_df['Median'],mode='lines',name=col_name+' (Median)'))
    fig.add_trace(go.Scatter(x=final_df.index,y=final_df['Standard deviation'],mode='lines',name=col_name+' (Standard deviation)'))
    fig.add_trace(go.Scatter(x=final_df.index,y=final_df['Variance'],mode='lines',name=col_name+' (Variance)'))
    fig.add_trace(go.Scatter(x=final_df.index,y=final_df['Mean absolute deviation'],mode='lines',name=col_name+' (Mean absolute deviation)'))
    fig.add_trace(go.Scatter(x=final_df.index,y=final_df['Count'],mode='lines',name=col_name+' (Count)'))
    fig.update_layout(
        title=col_name,
        xaxis_title="Date",
        yaxis_title="Values",
        legend_title="Statistical Parameters")
    fig.write_html("..\ShurovoData_2018H2\stastical_output\Plots\\"+col_name+".html")
    final_df.to_csv("..\ShurovoData_2018H2\stastical_output\Stats\\"+col_name+".csv",header=True)
    print("completed")
```

```python
for file in files:
    stastical_data(input_path+'\\'+file,file.split('.')[0])
```
