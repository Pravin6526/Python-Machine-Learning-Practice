```python
def stastical_data(path,col_name):
    print(col_name)
    df=pd.read_csv(path,skiprows=4,sep='\t')
    df.head()
    df.columns=[col_name]
    df.head()
    df.info()
    df.describe()
    df.isnull().sum()
    df.index=pd.to_datetime(df.index)
    df['date']=df.index.date
    final_df=df.groupby(['date',]).agg({col_name:['min','max','sum','mean','median','std','var','mad','count']})
    header = np.array(['Min','Max','Sum','Mean','Median', 'Standard deviation', 'Variance', 'Mean absolute deviation','Count'])
    final_df.columns=header
    final_df.to_csv('stastical_output/'+col_name+'.csv',header=True)
```
