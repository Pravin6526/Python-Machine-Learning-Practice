```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# legends

```python
x=np.linspace(0,10,11)
x
```

```python
y=x**2
y
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label='Graph X')
ax.plot(x,x**3,label='Graph X1')
ax.legend()
```

```python
#location of legends
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label='Graph X')
ax.plot(x,x**3,label='Graph X1')
ax.legend(loc=6)
```

```python
fig,ax=plt.subplots()
ax.plot(x,y,'b--')
ax.plot(x,x**3,'g--')
```
